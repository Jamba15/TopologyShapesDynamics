# Dirac Signal Processing on Simplicial Complexes
This notebook demonstrates  Dirac Signal Processing of nodes, edges and triangle signals on simplicial complexes. We will start by setting up the configuration and constructing a simplicial complex graph. Then, we will process signals based on the constructed graph.

```python
from functions_signalprocess import *
```

## Configuration and Graph Construction
We begin by parsing the configuration from a YAML file and constructing a simplicial complex graph based on the specified parameters. Look at the configuration file for further details.

```python
config = config_parser('signal_processing_NGF.yaml')
graph = SimplicialComplexConstructor(**config['simplicial_complex_constructor'])
graph.draw_graph()
```

## Dirac Signal Processing
We plot the inferred parameter m versus the true one and the normalized relative error as functions of time.
### Processing with D1
We will now process signals on dimensions 0-1. In the following section we will process the signal using D1

```python
process_with_dirac = 'D1'
if process_with_dirac == 'D1':
    _, D = graph.create_B1_D1()
else:
    _, D = graph.create_B2_D2()

signals = signal_elaboration(D)
signals.plot_results(save_fig=False)
```

### Processing with D2
Processing the signal on dimensions 1-2, using D2. Be aware that this only works if the simplicial complex has dimension 2 or higher.

```python
process_with_dirac = 'D2'
if process_with_dirac == 'D1':
    _, D = graph.create_B1_D1()
else:
    _, D = graph.create_B2_D2()

signals = signal_elaboration(D)
signals.plot_results(save_fig=False)
```
