# Triadic percolation on Poisson higher-order networks with triadic interactions. 


We consider Poisson higher-order networks with triadic interactions. Having a Poisson structural network with average degree `c`. The number of positive and negative regulator nodes displays a Poisson distribution with average degree `cp` and `cn` respectively. We first show the ordinary percolation on a Poisson network where triadic interactions are not present. The percolation displays a continuous second order phase transition. The dynamics is iterated until `t=Tmax` and the final `num` of points are collected to produce the figure.
The phase transition in absence of triadic interactions is shown in the following figure. The order parameter (the fraction of nodes in the giant component denoted as $R$) displays a continuous second order phase transition.

```python
config = config_parser('triadic.yml')
p_list, R_list = triadic_percolation_theory_poisson(config, "none")
plot_orbit_diagram(p_list, R_list)
```

When triadic interactions are present, the phase diagram becomes a orbit diagram and the order parameter (the fraction of nodes in the giant component denoted as $R$) displays period doubling and a route to chaos.

```python
config = config_parser('triadic_interaction_on.yml')
p_list, R_list = triadic_percolation_theory_poisson(config)
plot_orbit_diagram(p_list, R_list)
```

The time series of the dynamic is shown at a given $p$. The following figure shows a chaotic time series of the order parameter (the fraction of nodes in the giant component denoted as $R$) at $p=0.3$ (this can be changed in the configuration file named `triadic_in_time_fixed_p.yml`).

```python
config = config_parser('triadic_in_time_fixed_p.yml')
T_list, R_list = triadic_percolation_theory_poisson_time_series(config)
plot_time_series(T_list, R_list)
```
